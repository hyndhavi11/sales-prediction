**SALES PREDICTION USING PYTHON**

Sales prediction means predicting how much of a product, people will buy based on factors such as the amount you spend to advertise your product, the segment of people you advertise for, or the platform you are advertising on about your product.

First,import the required libraries:

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

Next,load the data:

```python
adv = pd.read_csv('/content/Advertising (1).csv')
```

```python
#shape
adv.shape
```

```python
#info of the data:
adv.info()
```

```python
#describing the data
adv.describe()
```

```python
#checking null values
adv.isnull().sum()
```

visualize the data:

```python
plt.figure(figsize=(8,6))
sns.histplot(adv['Sales'],kde=True)
plt.title('sales distribution')
plt.xlabel('Sales', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.show()
```

```python
#correlation heatmap
plt.figure(figsize=(6,4))
correlation_matrix = adv[['Sales', 'Radio', 'TV','Newspaper']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("correlation_matrix")
plt.show()
```

```python
#plotting between Tv vs Sales
px.scatter(adv, x='TV', y='Sales', width=850, height=500, title='Tv vs Sales')
```

```python
#plotting between Radio vs Sales
px.scatter(adv, x='Radio', y='Sales', width=850, height=500, title='Radio vs Sales')
```

```python
#plotting between Newspaper vs Sales
px.scatter(adv, x='Newspaper', y='Sales', width=850, height=500, title='Newspaper vs Sales')
```

Define the features and target variable:

```python
X = adv.drop('Sales', axis=1)
Y = adv['Sales']
print(X)
print(Y)
```

Split the data into traing and testing sets:

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=0)
```

```python
#Train the linear regression model
model = LinearRegression()
model.fit(X_train, Y_train)
```

prediction:

```python
Y_pred = model.predict(X_test)
Y_pred
```

```python
# actual prediction:
act_pred = pd.DataFrame({'Actual':Y_test.values.flatten(), 'Prediction':Y_pred.flatten()})
act_pred
```
